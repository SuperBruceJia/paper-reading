(a) $E(aX+bY) = aE(X)+bE(Y)$

(b) $Var(aX+bY) = a^2 Var(X) + b^2Var(Y) + 2abCov(X,Y)$, where $Cov(X,Y)$ is covariance of X and Y.

For simplicity, we can assume that $X$ and $Y$ are continuous random variables.


**Solution of the Q1.1 (a) - $E(aX+bY) = aE(X)+bE(Y)$**:<br>
***Proof.*** 
The expectation of a random variable $X$ is defined as:
\begin{equation}
    \begin{aligned}
        E(X) = \sum_{i=1}^{n} x_i p(x_i).
    \end{aligned}
\end{equation}
where $x_i$ are the values that $X$ can take on, and $p(x_i)$ is the probability that $X$ takes on the value $x_i$.

\begin{equation}
    \begin{aligned}
        E(aX + bY) & = \sum_{i=1}^{n} \sum_{j=1}^{m} (a x_i + b y_j) p(x_i, y_j) \\
        & = \sum_{i=1}^{n} \sum_{j=1}^{m} a x_i p(x_i, y_j) + \sum_{i=1}^{n} \sum_{j=1}^{m} b y_j p(x_i, y_j) \\
        & = \sum_{i=1}^{n} a x_i \sum_{j=1}^{m} p(x_i, y_j) + \sum_{j=1}^{m} b y_j \sum_{i=1}^{n} p(x_i, y_j) \\
        & = a \sum_{i=1}^{n} x_i p(x_i) + b \sum_{j=1}^{m} y_j p(y_j) \\
        & = a E(X) + b E(Y)
    \end{aligned}
\end{equation}
Thus, the equality has been proven.

**Solution of the Q1.1 (b) - $Var(aX+bY) = a^2 Var(X) + b^2Var(Y) + 2abCov(X,Y)$**:<br>
***Proof.*** We will first start with the definition of variance as follows,
\begin{equation}
    \begin{aligned}
        Var(aX + bY) & = E((aX + bY - E(aX + bY))^2)\\
        & = E((aX + bY - (aE(X) + bE(Y)))^2)\\
        & = E((a(X - E(X)) + b(Y - E(Y)))^2)\\
        & = E(a^2(X - E(X))^2 + 2ab(X - E(X))(Y - E(Y)) + b^2(Y - E(Y))^2)\\
        & = a^2E((X - E(X))^2) + 2abE((X - E(X))(Y - E(Y))) + b^2E((Y - E(Y))^2)\\
        & = a^2Var(X) + 2abCov(X, Y) + b^2Var(Y).
    \end{aligned}
\end{equation}
Thus, the equality has been proven.

**(c)** In this problem, we run stimulation experiments to compare the behavior between the usual variance estimator for indenpendent samples, and the variance estimator that accounts for correlation, derived in Q1.2 (b). In the code below, we generate correlated observation from $N(0, 1)$. Complete the estimator function `iid_estimator` and `corr_estimator` below for the usual variance estimator for indenpendent samples, and the variance estimator that accounts for correlation, and analyze the simulation results.

```python
# Coding part for Q 1.2(c)
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

# Usual variance estimator for independent samples
def iid_estimator(x):
    n = len(x)
    mu_hat = np.mean(x)
    sigma = np.sum((x - mu_hat)**2) / (n - 1)
    
    return sigma
    
# Variance estimator that accounts for correlation
def corr_estimator(x, R):
    n = len(x)
    mu = np.mean(x)
    cov_matrix_inv = np.linalg.inv(R)
    sigma = (x - mu).T @ cov_matrix_inv @ (x - mu) / n
    
    return sigma

n = 32
sigma = 1
n_sim = 1000
rho = 0.5
np.random.seed(101)

# Prepare the correlation matrix
R = np.ones((n,n)) * rho
np.fill_diagonal(R, 1)

# Run the simulation
sim = np.zeros((2, n_sim))
for i in range(n_sim):
    x = multivariate_normal.rvs(mean=np.zeros(n), cov=R * (sigma**2), size=1)
    sim[:, i] = [corr_estimator(x, R), iid_estimator(x)]

# Plot the result
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sim[0, :], color='gray', bins=30)
plt.xlabel('Estimate')
plt.title('Histogram of Estimates of $\\sigma^2$')
plt.axvline(x=sigma**2, color='red', linewidth=2)

plt.subplot(1, 2, 2)
plt.hist(sim[1, :], color='gray', bins=30)
plt.xlabel('Estimate')
plt.title('Histogram of Independent Estimates of $\\sigma^2$')
plt.axvline(x=sigma**2, color='red', linewidth=2)
```

**Analysis**:<br>
The variance estimator that accounts for correlation will typically be larger than the usual variance estimator for independent samples when there is a positive correlation between the elements of the vector $\mathbf{x}$. This is because a positive correlation tends to make the elements of $\mathbf{x}$ move together, leading to less variability in the sample.

Conversely, if there is a negative correlation between the elements of x, the variance estimator that accounts for the correlation may be smaller than the usual variance estimator. A negative correlation means that when one element of $\mathbf{x}$ is high, another tends to be low, which can reduce the overall variability in the sample.

In summary, accounting for correlation when estimating variance takes into consideration the dependencies between the elements of the vector, which can result in different estimates compared to the usual variance estimator for independent samples.

**Q1.3**:Let $(X, Y)$ be uniform on the unit square. Let $Z = X/Y$.
Find the probability density function (pdf) of $Z$.

**Hint:** pdf is the derivative of cumulative density function (cdf).

**Solution**:<br>
Using the hint that the pdf is the derivative of the cdf of Z, $F_Z(z)$ can be denoted as:
\begin{equation}
    \begin{aligned}
        F_Z(z) = P(Z \leq z) = P(\frac{X}Y \leq z).
    \end{aligned}
\end{equation}

The pdf $f_Z(z)$ is the derivative of the cdf with respect to $z$.
\begin{equation}
    \begin{aligned}
        f_Z(z) = \frac{d}{dz} P(Z \leq z).
    \end{aligned}
\end{equation}

- For $0 \leq z \leq 1$:
\begin{equation}
    \begin{aligned}
        P(Z \leq z) = P(X \leq zY) = \int_0^1 \int_0^{yz} \, dx \, dy = \int_0^1 yz \, dy = \frac{z}2.
    \end{aligned}
\end{equation}

- For $z > 1$:
\begin{equation}
    \begin{aligned}
        P(Z \leq z) = P(X/Y \leq z) & = P(X \leq zY)
        & = \int_0^{\frac{1}z} \int_0^{1} \, dx \, dy + \int_{\frac{1}z}^{1} \int_0^{yz} \, dx \, dy  = \int_0^1 yz \, dy = 1 - \frac{1}{2z}.
    \end{aligned}
\end{equation}

Now, differentiating for each region: \\
- For $0 \leq z \leq 1$:
\begin{equation}
    \begin{aligned}
        f_Z(z) = \frac{d}{dz} (\frac{z}2) = \frac{1}2.
    \end{aligned}
\end{equation}

- For $z > 1$:
\begin{equation}
    \begin{aligned}
        f_Z(z) = \frac{d}{dz} (1 - \frac{1}{2z}) = \frac{1}{2z^2}.
    \end{aligned}
\end{equation}

Thus, the PDF of $Z$ is:
\begin{equation}
    \begin{aligned}
        f_Z(z) =
          \begin{cases}
          \frac{1}{2} & \text{if } 0 < z \leq 1, \\
          \frac{1}{2z^2} & \text{if } z > 1, \\
          0 & \text{otherwise.}
          \end{cases}
    \end{aligned}
\end{equation}

**Q1.4** Let's assume that a binary classifier has a false positive rate of 0.01 (test positive but is actually negative), false negative rate of 0.3 (test negative but is actually positive). Assume that 20% of the samples are labled as postive. If one sample were predicted as negative by the model, what are the likelihood that the sample is labeled as postive? (Show your steps)

**Solution**:<br>
To calculate the likelihood that a sample predicted as negative by the model is actually labeled as positive, we can use Bayes' theorem. We are looking for the conditional probability that a sample is labeled as positive given that it was predicted as negative. Let's denote:

- P(Positive): the probability that a sample is labeled as positive (prior probability), which is 0.20 (20% of the samples are labeled as positive).
- P(Negative): the probability that a sample is labeled as negative, which is 1 - P(Positive), so P(Negative) = 1 - 0.20 = 0.80.
- P(Negative|Positive): the probability of being predicted as negative given that the sample is actually positive, which is the false negative rate, so P(Negative|Positive) = 0.30.
- P(Negative|Negative): the probability of being predicted as negative given that the sample is actually negative, which is 1 minus the false positive rate, so P(Negative|Negative) = 1 - 0.01 = 0.99.

Now, we want to find P(Positive|Negative), the probability that a sample is labeled as positive given that it was predicted as negative:

By Bayes' theorem:

\begin{equation}
    \begin{aligned}
        P(\text{Positive}|\text{Negative}) = \frac{P(\text{Negative}|\text{Positive}) \times P(\text{Positive})}{P(\text{Negative}|\text{Positive}) \times P(\text{Positive}) + P(\text{Negative}|\text{Negative}) \times P(\text{Negative})}.
    \end{aligned}
\end{equation}

Substituting the values:

\begin{equation}
    \begin{aligned}
        P(\text{Positive}|\text{Negative}) = \frac{0.30 \times 0.20}{0.3 \times 0.20 + 0.99 \times 0.80} = 0.0704.
    \end{aligned}
\end{equation}

So, the likelihood that a sample predicted as negative by the model is actually labeled as positive is 7.04%.

In your previous machine learning courses, you've likely encountered various types of probability distributions like Gaussian, Poisson, and Bernoulli. You may not know that these distributions are part of a larger, unified framework known as the "exponential family" of distributions. Understanding the exponential family is crucial for advancing machine learning and statistics because it provides a general way to express a wide range of probability distributions using a standard form. This unified representation makes it easier to develop algorithms and statistical methods that can work across different types of data.



The exponential family of distributions can be expressed in a canonical form as follows:

\begin{equation}
p({\mathbf{x}} ; \boldsymbol{\theta}) = h(\mathbf{x}) \exp \left( \boldsymbol{\eta}(\boldsymbol{\theta}) \cdot \mathbf{T}(\mathbf{x}) - A(\boldsymbol{\theta}) \right)
\end{equation}

Here's what each term represents:


     
- $\mathbf{x}$ is the observed data, which can be potentially high-dimensional vector

- $ \boldsymbol{\theta} $ are the natural parameters of the distribution, which is a vector
- $ h(\mathbf{x}) $ is the base measure, which is a function of the data $ \mathbf{x} $.
- $ \boldsymbol{\eta}(\boldsymbol{\theta}) $ is a function mapping the natural parameters $ \boldsymbol{\theta} $ to the canonical parameters.
- $ \mathbf{T}(\mathbf{x}) $ is the sufficient statistic, which is a function of the data $ \mathbf{x} $. In the machine learning community, $\mathbf{T}(\mathbf{x})$ is sometimes called the feature vector of the distribution.
- $ A(\boldsymbol{\theta}) $ is the log-partition function, which ensures that the distribution integrates with 1.
- $\boldsymbol{\eta}(\boldsymbol{\theta}) \cdot \mathbf{T}(\mathbf{x})$ denotes the inner product between the two vectors,  $\mathbf{T(x)}$ and $\boldsymbol{\eta}(\boldsymbol{\theta})$



The beauty of this formulation is its generality. By choosing different forms for $ h(\mathbf{x}) $, $ \boldsymbol{\eta}(\boldsymbol{\theta}) $, $ \mathbf{T}(\mathbf{x}) $, and $ A(\boldsymbol{\theta}) $, you can derive a wide range of specific distributions like Gaussian, Poisson, and Bernoulli.

The expression may look scary at the beginning. Let us start with a probability density you are the most familiar with and make it look like a member of the exponential family: univariate Gaussian distribution!

**Q2.3 (Nice properties of the partition function)**: The partition function is the normalizer of the pdf (probability density function).

\begin{equation*}
    p(\mathbf{x}; \boldsymbol{\eta}) \propto h(\mathbf{x}) \exp \left( \boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}) \right).
\end{equation*}
The partition function ensures that pdf integrates into one. In other words, the partition function makes the proportional sign to the equality.
\begin{equation*}
    p(\mathbf{x}; \boldsymbol{\eta}) = \frac{1}{e^{A(\boldsymbol{\eta})}} h(\mathbf{x}) \exp \left( \boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}) \right).
\end{equation*}

\begin{equation*}
  e^{A(\boldsymbol{\eta})} = \int h(\mathbf{x}) \exp \left( \boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}) \right) d\mathbf{x} \quad \rightarrow \quad A(\boldsymbol{\eta}) = \log \int h(\mathbf{x}) \exp \left( \boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}) \right) d\mathbf{x}
\end{equation*}

-- Show that the gradient of the partition function has the following property:
\begin{equation*}
  \frac{\partial A}{ \partial \boldsymbol{\eta}^T} = \mathbb{E} \left[ \mathbf{T}(\mathbf{x}) \right],
\end{equation*}
where $\mathbb{E}$ denotes the expectation.

**Solution**:<br>
let's differentiate $A(\boldsymbol{\eta})$ with respect to $\boldsymbol{\eta}^T$:
\begin{equation}
    \begin{aligned}
        \frac{\partial A}{\partial \boldsymbol{\eta}^T} = \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\log \int h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) \, d\mathbf{x}\right).
    \end{aligned}
\end{equation}

Using the chain rule, we get:
\begin{equation}
    \begin{aligned}
        \frac{\partial A}{\partial \boldsymbol{\eta}^T} = \frac{1}{\int h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) \, d\mathbf{x}} \cdot \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\int h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) \, d\mathbf{x}\right).
    \end{aligned}
\end{equation}

Now, let's focus on the second term, which is the derivative of the integral:
\begin{equation}
    \begin{aligned}
        \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\int h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) \, d\mathbf{x}\right).
    \end{aligned}
\end{equation}

By the Leibniz rule for differentiating under the integral sign, we can exchange the derivative and the integral:
\begin{equation}
    \begin{aligned}
        \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\int h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) \, d\mathbf{x}\right) = \int \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}))\right) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

Now, we need to differentiate $h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}))$ with respect to $\boldsymbol{\eta}^T$. This is straightforward because $h(\mathbf{x})$ does not depend on $\boldsymbol{\eta}$, and we can use the chain rule for the exponential term:
\begin{equation}
    \begin{aligned}
        \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(h(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}))\right) = h(\mathbf{x}) \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}))\right).
    \end{aligned}
\end{equation}

Now, we can find the derivative of the exponential term with respect to $\boldsymbol{\eta}^T$:
\begin{equation}
    \begin{aligned}
        \frac{\partial}{\partial \boldsymbol{\eta}^T}\left(\exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x}))\right) = \mathbf{T}(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})).
    \end{aligned}
\end{equation}

Now, substitute this result back into the integral expression:
\begin{equation}
    \begin{aligned}
        \int \mathbf{T}(\mathbf{x}) \exp(\boldsymbol{\eta} \cdot \mathbf{T}(\mathbf{x})) h(\mathbf{x}) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

This expression is exactly the expected value (mean) of $\mathbf{T}(\mathbf{x})$ with respect to the distribution $p(\mathbf{x}; \boldsymbol{\eta})$:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \int \mathbf{T}(\mathbf{x}) p(\mathbf{x}; \boldsymbol{\eta}) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

So, we have shown that:
\begin{equation}
    \begin{aligned}
        \frac{\partial A}{\partial \boldsymbol{\eta}^T} = \mathbb{E}[\mathbf{T}(\mathbf{x})].
    \end{aligned}
\end{equation}

The equality has been proven.

**Q2.4 (Maximum Likelihood (ML) using exponential family)**: Now, let's do Maximum Likelihood estimation with the exponential family. Let's assume that $\mathcal{D} = \{ \mathbf{x_1}, \cdots, \mathbf{x_N} \}$ denotes the training data of size $N$. The ML estimation maxmizes the following loss function:

\begin{eqnarray*}
 \hat{\boldsymbol{\eta}} = \arg \max_{\boldsymbol{\eta}} \ell( \boldsymbol{\eta}; \mathcal{D}),
\end{eqnarray*}
where
\begin{eqnarray*}
\ell(\boldsymbol{\eta}; \mathcal{D}) = \sum_{n=1}^{N} \log p(\mathbf{x_n} ; \boldsymbol{\eta})
\end{eqnarray*}

**Hint:** you can drop $h(\mathbf{x})$ term in the loss function because it does not depend on $\boldsymbol{\eta}$ and does not contribute to the optimization. You should use the results from Q2.3.

--  Show that the optimal $\boldsymbol{\eta}$ satisfies the following condition:
\begin{eqnarray*}
    \mathbb{E} \left[ \mathbf{T} (\mathbf{x})\right] = \dfrac{1}{N} \sum_{n=1}^{N} \mathbf{T}(\mathbf{x_n})
\end{eqnarray*}

**Solution**:<br>
The MLE seeks to maximize the log-likelihood function:
\begin{equation}
    \begin{aligned}
        \boldsymbol{\hat{\eta}} = \arg \max_{\boldsymbol{\eta}} \ell(\boldsymbol{\eta}; \mathcal{D}) = \arg \max_{\boldsymbol{\eta}} \sum_{n=1}^{N} \log p(\mathbf{x}_n ; \boldsymbol{\eta}).
    \end{aligned}
\end{equation}

Let's consider the derivative of the log-likelihood function with respect to $\boldsymbol{\eta}$:
\begin{equation}
    \begin{aligned}
        \frac{\partial \ell}{\partial \boldsymbol{\eta}} = \sum_{n=1}^{N} \frac{\partial}{\partial \boldsymbol{\eta}} \log p(\mathbf{x}_n ; \boldsymbol{\eta}).
    \end{aligned}
\end{equation}

Now, let's focus on the expectation of the sufficient statistics:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \int \mathbf{T}(\mathbf{x}) p(\mathbf{x} ; \boldsymbol{\eta}) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

For simplicity, let's denote $\mathbf{T}_n = \mathbf{T}(\mathbf{x}_n)$ as the sufficient statistics for the $n$-th data point. Now, consider the gradient of the log-likelihood with respect to $\boldsymbol{\eta}$ and evaluate it at the optimal $\boldsymbol{\hat{\eta}}$:
\begin{equation}
    \begin{aligned}
        \left.\frac{\partial \ell}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = \sum_{n=1}^{N} \left.\frac{\partial}{\partial \boldsymbol{\eta}} \log p(\mathbf{x}_n ; \boldsymbol{\eta})\right|_{\boldsymbol{\hat{\eta}}} = \sum_{n=1}^{N} \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\eta})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\eta})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}}.
    \end{aligned}
\end{equation}

Now, set this gradient to zero to find the optimal $\boldsymbol{\hat{\eta}}$:
\begin{equation}
    \begin{aligned}
        \left.\frac{\partial \ell}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

This implies that at the MLE, the gradient of the log-likelihood with respect to $\boldsymbol{\eta}$ is zero. Now, consider the expectation of the sufficient statistics:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \int \mathbf{T}(\mathbf{x}) p(\mathbf{x} ; \boldsymbol{\eta}) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

At the MLE $\boldsymbol{\hat{\eta}}$, the log-likelihood is maximized, which means that the gradient is zero. In other words:
\begin{equation}
    \begin{aligned}
        \left.\frac{\partial \ell}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

Now, consider the gradient of the log-likelihood with respect to $\boldsymbol{\eta}$ at the MLE:
\begin{equation}
    \begin{aligned}
        \left.\frac{\partial \ell}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = \sum_{n=1}^{N} \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

Since the gradient is zero, we have:
\begin{equation}
    \begin{aligned}
        \sum_{n=1}^{N} \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

Now, consider the term inside the summation:
\begin{equation}
    \begin{aligned}
        \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = \frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \cdot 0 = 0.
    \end{aligned}
\end{equation}

This implies that each term in the summation is zero:
\begin{equation}
    \begin{aligned}
        \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

Since each term is zero, it follows that the entire sum is zero:
\begin{equation}
    \begin{aligned}
        \sum_{n=1}^{N} \left.\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}}\right|_{\boldsymbol{\hat{\eta}}} = 0.
    \end{aligned}
\end{equation}

Now, let's return to the expectation of the sufficient statistics:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \int \mathbf{T}(\mathbf{x}) p(\mathbf{x} ; \boldsymbol{\hat{\eta}}) \, d\mathbf{x}.
    \end{aligned}
\end{equation}

Since each term in the summation is zero, the expectation becomes:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \sum_{n=1}^{N} \frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \int \mathbf{T}(\mathbf{x}) \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}{\partial \boldsymbol{\eta}} \, d\mathbf{x}.
    \end{aligned}
\end{equation}

Now, notice that $\int \mathbf{T}(\mathbf{x}) \frac{\partial p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} {\partial \boldsymbol{\eta}} \, d\mathbf{x}$ is the same as $\mathbf{T}(\mathbf{x}_n)$ because the derivative of the probability distribution with respect to $\boldsymbol{\eta}$ evaluated at the MLE $\boldsymbol{\hat{\eta}}$ is zero.

Therefore, we have:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \sum_{n=1}^{N} \frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \mathbf{T}(\mathbf{x}_n).
    \end{aligned}
\end{equation}

Now, notice that $p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})$ is the likelihood of data point $\mathbf{x}_n$ given the MLE $\boldsymbol{\hat{\eta}}$, so it's a constant. Therefore, we can factor it out of the sum:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \sum_{n=1}^{N} \mathbf{T}(\mathbf{x}_n).
    \end{aligned}
\end{equation}

Now, since each term in the summation is multiplied by the constant $\frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})}$, it follows that:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \frac{1}{p(\mathbf{x}_n ; \boldsymbol{\hat{\eta}})} \cdot N \cdot \frac{1}{N} \sum_{n=1}^{N} \mathbf{T}(\mathbf{x}_n).
    \end{aligned}
\end{equation}

Finally, this leads to the desired result:
\begin{equation}
    \begin{aligned}
        \mathbb{E}[\mathbf{T}(\mathbf{x})] = \frac{1}{N} \sum_{n=1}^{N} \mathbf{T}(\mathbf{x}_n).
    \end{aligned}
\end{equation}

The equality has been proven.

### Problem 3: Multi-class Logistic Regression (20 points)

In the problem, we want to solve the multiclass classification problem with logistic regression.

-The matrix $\mathbf{z}_i$ prepresents the inner product such that $\mathbf{z}_i=\left[ w_1^T \mathbf{x}_i , w_2^T \mathbf{x}_i, \cdots, w_K^T \mathbf{x}_i \right]$. We use $z_{ik}$ to denote the $k$ element of $\mathbf{z}_i$




**Q3.1**: Recall that in multiclass regression, the probability is determined by taking the softmax of the inner product $Z_i$. Given the above definitions, please write down the log-likelihood function and simplify:

**Hint 1:** $\mathbb{P}(y_i=k; \mathbf{W}| \mathbf{x}_i ) \propto e^{z_{ik}}$ and you need to normalize this to get the correct probability.

**Hint 2:** The log likelihood of the sample $\mathcal{L}( \mathbf{x}_i,y_i ; \mathbf{W}) = \sum_{k=1}^{K} \log \mathbb{P}(y_i=k; \mathbf{W} | \mathbf{x}_i)$

**Solution**:<br>
The probability for each data point $\mathbf{x}_i$ to belong to class $k$ can be calculated using the softmax function:
\begin{equation}
    \begin{aligned}
        \mathbb{P}(y_i = k | \mathbf{x}_i) = \frac{e^{z_{ik}}}{\sum_{j=1}^{K} e^{z_{ij}}},
    \end{aligned}
\end{equation}
where $\mathbb{P}(y_i = k | \mathbf{x}_i)$ is the probability of data point $\mathbf{x}_i$ belonging to class $k$.

The log-likelihood function for the entire dataset is the product of the probabilities of individual data points. Assuming that each data point is independent, the log-likelihood is:
\begin{equation}
    \begin{aligned}
        \mathcal{L}(\mathbf{w}) = \sum_{k=1}^K \log \mathbb{P}(y_i = k | \mathbf{x}_i),
    \end{aligned}
\end{equation}
where $\mathbf{w}$ represents the parameter matrix containing all the weight vectors $\mathbf{w}_k$.

Now, let's substitute the expression for $\mathbb{P}(y_i = k | \mathbf{x}_i)$ into the log-likelihood:
\begin{equation}
    \begin{aligned}
        \mathcal{L}(\mathbf{w}) = \sum_{k=1}^K \log \left( \frac{e^{z_{ik}}}{\sum_{j=1}^K e^{z_{ij}}} \right).
    \end{aligned}
\end{equation}

So, the log-likelihood function for multi-class logistic regression is:
\begin{equation}
    \begin{aligned}
        \mathcal{L}(\mathbf{w}) = \sum_{k=1}^K \left( z_{ik} - \log \sum_{j=1}^K e^{z_{ij}} \right).
    \end{aligned}
\end{equation}

\begin{equation}
    \begin{aligned}
        P(y_i = k | \mathbf{X}_i) = \frac{e^{z_{ik}}}{\sum_{j=1}^{K} e^{z_{ij}}}.
    \end{aligned}
\end{equation}

Now, let's consider the modified likelihood function after subtracting $a$ from $Z_i$:
\begin{equation}
    \begin{aligned}
        P(y_i = k | \mathbf{X}_i) = \frac{e^{z_{ik} - a}}{\sum_{j=1}^{K} e^{z_{ij} - a}}.
    \end{aligned}
\end{equation}

The key insight here is that we are applying the same constant $a$ to the numerator and denominator. This constant term does not change the relative values of $e^{z_{ik}}$ and $e^{z_{ij}}$ for any $k$ and $j$. In other words, it doesn't change the ordering of the probabilities. Therefore, it does not affect the class with the highest probability, and it does not affect the predicted class.

When substituting this relationship into your simplified log-likelihood, it ensures that the greatest value is zero even if other values would underflow. This workaround leads to relatively decent preformance even in the case of a large $X_i$. While you may have encountered this before, it is explicitly known as the LogSumExp trick and it is an important substitution for implementing a multiclass regression classification model.

**Q3.4:** You may remember that a regularization term, commonly denoted as $R(f)$, is often added to the loss function in classification problem. When tuning the model for optimal preformance, $R(f)$ is usually scaled by some constant $\lambda$ which controls the effect of the regularization on the outcome of the loss function. Why do we add and tune this term to our loss? How can it improve our model's preformance?

**Solution**:<br>
(1) **Why do we add and tune this term to our loss?**<br>
Common regularization techniques include L1 regularization (Lasso), L2 regularization (Ridge), and their combination (Elastic Net). Each of these techniques penalizes the model's weights differently, leading to different effects on model complexity and feature selection. Overall, it helps improve model generalization by preventing overfitting, encouraging simpler models, and providing control over the bias-variance trade-off.

(2) **How can it improve our model's performance?**<br>
1. **Preventing Overfitting:** Overfitting occurs when a model learns to fit the training data too closely, capturing noise and idiosyncrasies in the data rather than the underlying patterns. Regularization helps prevent overfitting by adding a penalty term to the loss function that discourages the model from learning overly complex or flexible representations. This encourages the model to generalize better to unseen data.

2. **Smoother Decision Boundaries:** Regularization encourages the model to have smoother and simpler decision boundaries. In the context of classification, this means that the model is less likely to make erratic decisions that fit the training data perfectly but do not generalize well. Smoother decision boundaries lead to better generalization.

3. **Improved Numerical Stability:** Regularization can improve the numerical stability of the training process. It helps to mitigate issues like numerical instability in optimization algorithms, especially when dealing with ill-conditioned or nearly singular data.

4. **Feature Selection:** Some regularization techniques, like L1 regularization (Lasso), encourage sparsity in the model's weights. This can effectively perform feature selection by pushing the weights of irrelevant features towards zero. It simplifies the model and can lead to a more interpretable and efficient model.

5. **Reducing Model Complexity:** By penalizing large weight values, regularization encourages the model to use fewer features and focus on the most important ones. This can reduce the complexity of the model, making it less prone to overfitting.

6. **Balancing Bias and Variance:** Regularization allows you to control the bias-variance trade-off in your model. By adjusting the regularization strength $\lambda$, you can find the right balance between fitting the training data well and preventing overfitting. Smaller $\lambda$ values lead to less regularization and potentially higher variance but lower bias, while larger $\lambda$ values increase bias but reduce variance.

```python
# Training and Evaluation of Multi Class Logistic Model Performance on synthetic dataset (dataset.pkl)
import pickle
import numpy as np
import pandas as pd
from scipy.special import softmax
from copy import deepcopy
import matplotlib.pyplot as plt

def permute_data(X, y):
    """ Randomly permute data """
    perm = np.random.permutation(len(X))
    return X[perm], y[perm]


def softmax_stable(Z, axis=1):
    ## <! code required> Optional: You can prevent Overflow by subtracting the inputs with the maximum value in the input set.
    exp_logits = np.exp(Z - np.max(Z, axis=axis, keepdims=True))
    out = exp_logits / np.sum(exp_logits, axis=axis, keepdims=True)
    
    return out


def logistic_regression_grad(X, y_onehot, w, reg_lambda):
    """
    y_onehot: onehot encoded labels
    """
    m = X.shape[0]
    Z = X @ w
    P = softmax_stable(Z, axis=1)
    grad = (1/m) * X.T @ (P - y_onehot) + reg_lambda * w
    
    return grad
    

def logistic_regression_loss(X, y_onehot, w, reg_lambda):
    m = X.shape[0]
    Z = X @ w
    P = softmax_stable(Z, axis=1)
    loss = -(1/m) * np.sum(y_onehot * np.log(P + 1e-8)) + (reg_lambda/2) * np.sum(w**2)  # added epsilon to prevent log(0)

    return loss


def multiclass_accuracy(X, y_onehot, w):
    Z =  X @ w
    P = softmax_stable(Z, axis=1)
    predictions = np.argmax(P, axis=1)
    true_labels = np.argmax(y_onehot, axis=1)
    return np.mean(predictions == true_labels)


def training_multiclass_reg(X_train, y_train, X_validation, y_validation, w, STEP_SIZE, reg_lambda):
    BATCH_SIZE = 2
    NUM_TRAIN = len(X_train)
    X_p, y_p = permute_data(X_train, y_train)
    epoch_loss_history = []

    for epoch in range(NUM_EPOCHS):
        epoch_loss = AverageMeter()
        for i in range(0, NUM_TRAIN, BATCH_SIZE):
            X_batch = X_p[i: min(i + BATCH_SIZE, NUM_TRAIN)]
            y_batch = y_p[i: min(i + BATCH_SIZE, NUM_TRAIN)]
            
            loss = logistic_regression_loss(X_batch, y_batch, w, reg_lambda)
            epoch_loss.update(loss)
            grad = logistic_regression_grad(X_batch, y_batch, w, reg_lambda)
            
            ## <! code required>
            w -= STEP_SIZE * grad
            
        epoch_loss_history.append(epoch_loss.avg)
        X_p, y_p = permute_data(X_train, y_train)
    return w, epoch_loss_history


class AverageMeter:
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count



# Load Dataset
with open("dataset.pkl", 'rb') as f:
    dataset = pickle.load(f)


# Parameters
NUM_EPOCHS = 500
STEP_SIZE = 0.01
FEAT_SIZE = dataset['x_train'].shape[1]
NUM_CLASSES = dataset['y_train'].shape[1]
INIT_W = np.random.randn(FEAT_SIZE, NUM_CLASSES)

# Parameters for regularization (lambda_values)
LAMBDA1, LAMBDA2 = 1e-4, 1e-2

# Training and Evaluation
# Re-training with LAMBDA1 (1e-4)
w_final_1, epoch_loss_1 = training_multiclass_reg(dataset['x_train'], dataset['y_train'],
                                                  dataset['x_test'], dataset['y_test'],
                                                  deepcopy(INIT_W), STEP_SIZE, LAMBDA1)

val_loss_1 = logistic_regression_loss(dataset['x_test'], dataset['y_test'], w_final_1, LAMBDA1)
val_acc_1 = multiclass_accuracy(dataset['x_test'], dataset['y_test'], w_final_1)

# Re-training with LAMBDA2 (1e-2)
w_final_2, epoch_loss_2 = training_multiclass_reg(dataset['x_train'], dataset['y_train'],
                                                  dataset['x_test'], dataset['y_test'],
                                                  deepcopy(INIT_W), STEP_SIZE, LAMBDA2)

val_loss_2 = logistic_regression_loss(dataset['x_test'], dataset['y_test'], w_final_2, LAMBDA2)
val_acc_2 = multiclass_accuracy(dataset['x_test'], dataset['y_test'], w_final_2)

print(val_loss_1, val_acc_1, val_loss_2, val_acc_2)
```
